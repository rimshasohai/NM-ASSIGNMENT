```python
import numpy as np
#standard RK method 


#EXACT SOLUTION
#def ode(x,y):
    #return ((1.33-1)/(51+1.33))*(x*y)

    #AFTER INTEGRATING IT WITH RESPECT TO dx
    #return  ((1.33-1)/(51+1.33))*((x**2/2)*y)
    
#def StandardMethod(a,b,c1,c2,c3,c4,h):
#    L1=[]
#    Yi_1=(b-1)/b
#    #print(Yi_1)
#    xi_1=0
#    theta=0.5
#    while xi_1<=1.2961:         #domain#0<x<1.2961((a*b)/(a+b))
#        K1=h*(ode(xi_1,Yi_1))
#        K2=h*(ode((xi_1+h/2),(Yi_1+K1/2)))
#        K3=h*(ode((xi_1+h/2),(Yi_1+K2/2)))
#        K4=h*(ode((xi_1+h),(Yi_1+K3)))
#        K=(c1*K1+c2*K2+c3*K3+c4*K4)/6
#        Yi=Yi_1+K           #reading at xi
        #print(Yi)
#        Yi_1=Yi
#        L1.append(Yi_1)
        #print(x)
#        xi_1=xi_1+h         #for next step
        
#    return L1

#============================================
#NUMERICAL SOLUTION

def ode(x,y):

    #AFTER INTEGRATING IT WITH RESPECT TO dx
    #return  ((1.33-1)/(51+1.33))*((x**2/2)*y)
    
    return  ((1.33-1)/(51+1.33))*((1/x)*ln(y))
    
def StandardMethod(a,b,c1,c2,c3,c4,h):
    L1=[]
    Yi_1=(b-1)/b
    #print(Yi_1)
    xi_1=0
    theta=0.5
    while xi_1<=1.2961:         #domain#0<x<1.2961((a*b)/(a+b))
        K1=h*(ode(xi_1,Yi_1))
        K2=h*(ode((xi_1+h/2),(Yi_1+K1/2)))
        K3=h*(ode((xi_1+h/2),(Yi_1+K2/2)))
        K4=h*(ode((xi_1+h),(Yi_1+K3)))
        K=(c1*K1+c2*K2+c3*K3+c4*K4)/6
        Yi=Yi_1+K           #reading at xi
        #print(Yi)
        Yi_1=Yi
        L1.append(Yi_1)
        #print(x)
        xi_1=xi_1+h         #for next step
        
    return L1

#============================================
#modified RK Mwthod
def ModifiedMethod(a,b,c1,c2,c3,c4,h):
    Yi_1=(b-1)/b
    L2=[]
    #print(Yi_1)
    xi_1=0
    theta0=0
    theta1=0.7
    theta2=0.4
    theta3=1
    c0=1
    k0=0
    
    while xi_1<=1.2961: #last two values of roll num
        K1=h*(ode(xi_1+theta0*h,(Yi_1+k0/c0)))    #w0=1/c0
        K2=h*(ode((xi_1+theta1*h),(Yi_1+K1/c1)))  #w1=1/c1
        K3=h*(ode((xi_1+theta2*h),(Yi_1+K2/c2)))  #w2=1/c2
        K4=h*(ode((xi_1+theta3*h),(Yi_1+K3/c3)))  #w3=1/c3
        sum_of_coefficients=c1+c2+c3+c4
        K=(c1*K1+c2*K2+c3*K3+c4*K4)/sum_of_coefficients
        #print('THE VALUE OF K:',K)
        
        Yi=Yi_1+K
        #print('USING Yi_1 TO COMPUTE Yi:',Yi)
        Yi_1=Yi
        L2.append(Yi_1)
        #print("sum of coeff",sum_of_coefficients)
        xi_1=xi_1+h
        
   
    return L2
    

#==============================================


#Calculation of root mean square error
s=[1,2,4,8,16]
eList=[]      #error list     
for loop in s:
    L1=StandardMethod(22,1.33,1,2,2,1,0.5/loop)
    L2=ModifiedMethod(22,1.33,2,1,2,2,0.5/loop)
    print("OUTPUT VALUES OF Yi AT Xi from Standard Method:",L1)
    print()
    print("OUTPUT VALUES OF Yi AT Xi from Modified Method:",L2)
    print()
    result=np.subtract(L1,L2)
    result2=np.square(result)
    a=sum(result2)
    a=a/len(result2)
    import math
    error=math.sqrt(a)        #root mean square error 
    
    print("Root mean square error for h=0.5/{} is,".format(loop), error)
    print()
    eList.append(error)
    
    
import matplotlib.pyplot as plt
h = [0.5,0.25,0.125,0.0625,0.03125]     #calculated root mean square error for these values of step size


  
# plotting the points 
plt.plot(h, eList)
  
# naming the x axis
plt.xlabel('h')
# naming the y axis
plt.ylabel('RMS error')
  
# giving a title to my graph
plt.title('Graph')
  
# function to show the plot
plt.show()
#==========================
#h is the step size
#a and b will be any number to caculate base value or initial value of y
#c1,c2,c3,c4 are coefficients of K
```
